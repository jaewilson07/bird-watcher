## Load environment and secrets
We use a .env file located in the root folder to store project configuration settings and secrets that we don't want the world to know.

An example could be my password to a website or the IP address of my internet phone. 

I don't want to store that on the internet or accidentally share that information with you when I give you a copy of this project.  You need to set your own .env values! Use the .env_sample file.

```python
from dotenv import load_dotenv
import os

PHONE_IP =os.environ.get("PHONE_IP")

if PHONE_IP:
    print(f"PHONE_IP: {PHONE_IP}")
else:
    print("PHONE_IP environment variable is not set.")


load_dotenv()  # Load environment variables from .env file

PHONE_IP = os.environ.get("PHONE_IP")
if PHONE_IP:
    print(f"PHONE_IP: {PHONE_IP}")
else:
    print("PHONE_IP environment variable is not set.")
```
